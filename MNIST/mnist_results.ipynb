```python
### Imports
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torchvision
from torchvision.datasets import MNIST
from torchvision import transforms
```

```python
# Define a transformer
transform = transforms.ToTensor()

# Download the data
# train data: training to be done on these
train_data = MNIST(root = "/home/aakavan/git_projects/practice_projects/MNIST/data", 
                  train = True, 
                  download = True,
                  transform = transform)


# test data: no training to be done on these
test_data = MNIST(root = "/home/aakavan/git_projects/practice_projects/MNIST/data",
                  train = False,
                  download = True,
                  transform = transform)

print(len(train_data), len(test_data))
```
