```python
### Imports
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torchvision
from torchvision.datasets import MNIST
from torchvision import transforms
from torch.utils.data import DataLoader
import torch.nn as nn
```

```python
# Define a transformer
transform = transforms.ToTensor()

# Download the data
# train data: training to be done on these
train_data = MNIST(root = "/home/aakavan/git_projects/practice_projects/MNIST/data", 
                  train = True, 
                  download = True,
                  transform = transform)


# test data: no training to be done on these
test_data = MNIST(root = "/home/aakavan/git_projects/practice_projects/MNIST/data",
                  train = False,
                  download = True,
                  transform = transform)

print(len(train_data), len(test_data))
```

```python
random_index = np.random.randint(0, len(train_data))

img, label = train_data[random_index]

plt.imshow(img.squeeze(), cmap="gray")
```

```python
# Now we need to make data loaders
# Train loader
train_loader = DataLoader(
    train_data, 
    batch_size=64,
    shuffle=True
)

# test loader
test_loader = DataLoader(
    test_data,
    batch_size=64,
    shuffle=False
)

# Seeing these batches
images, labels = next(iter(train_loader))
print(images.shape, labels.shape)
```
